```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/lichess_db_standard_rated_2013-01.csv')
df.replace('?', np.nan, inplace=True)
df = df.dropna()
display(df)
```

```python

target = df["selected_move"]
labels  = df.drop(columns=["selected_move"])

print(f"DataFrame shape: {df.shape}")
print(f"Target shape: {df['selected_move'].shape}")
print(f"Labels shape: {df.drop(columns=['selected_move']).shape}")
```

## Split the data

```python
from sklearn.model_selection import train_test_split

# train test val split
train_x, test_x, train_y, test_y = train_test_split(labels, target, test_size=0.1, random_state=42)

print(f"Train size: {train_x.shape}")
print(f"Test size: {test_x.shape}")

val_x, test_x, val_y,  test_y = train_test_split(test_x, test_y, test_size=0.5, random_state=42)

print(f"Train size: {len(train_x)}")
print(f"Val size: {len(val_x)}")
print(f"Test size: {len(test_x)}")

train_x.describe()
```

# Pre-process the data

```python
# from sklearn.imputer import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Pre-process the data
numerical_cols   = ["white_elo", "black_elo"]
binary_cols = ["blacks_move"]
categorical_cols = [col for col in train_x.columns if col not in numerical_cols + binary_cols]

print(f"Numerical columns: {numerical_cols}")
print(f"Binary columns: {binary_cols}")
print(f"Categorical columns: {categorical_cols}")

# Since we dropped NA values earlier, we can directly scale
# Create and fit the scaler on training data
numerical_transformer = StandardScaler()
numerical_transformer.fit(train_x[numerical_cols])

# Transform all datasets using the fitted scaler
train_x[numerical_cols] = numerical_transformer.transform(train_x[numerical_cols])
val_x[numerical_cols]   = numerical_transformer.transform(val_x[numerical_cols])
test_x[numerical_cols]  = numerical_transformer.transform(test_x[numerical_cols])

print("Numerical columns have been standardized (mean=0, std=1)")
print(f"Train numerical stats after scaling:")
print(train_x[numerical_cols].describe())


# Turn categorical features into a one-hot encoding
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
categorical_transformer.fit(train_x[categorical_cols])
train_cat = categorical_transformer.transform(train_x[categorical_cols])
val_cat   = categorical_transformer.transform(val_x[categorical_cols])
test_cat  = categorical_transformer.transform(test_x[categorical_cols])
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10, random_state=42, n_jobs=-1)
model.fit(train_cat, train_y)  
```

```python
# score the model
train_acc = model.score(train_cat, train_y)
val_acc   = model.score(val_cat, val_y)
print(f"Train accuracy: {train_acc:.4f}")
print(f"Val accuracy: {val_acc:.4f}")
```
